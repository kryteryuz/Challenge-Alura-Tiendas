### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
print("\nAnálisis de Facturación Total por Tienda")

for nombre, df in [("Tienda 1", tienda), ("Tienda 2", tienda2), ("Tienda 3", tienda3), ("Tienda 4", tienda4)]:
    try:
        total = df["Precio"].sum()
        print(f"{nombre}: $ {total:,.2f}")
    except Exception as e:
        print(f"Hubo un error en {nombre}: {e}")
```

# 2. Ventas por categoría

```python
print("\n2. Ventas por Categoría de cada Tienda")

for nombre, df in [("Tienda 1", tienda), ("Tienda 2", tienda2), ("Tienda 3", tienda3), ("Tienda 4", tienda4)]:
    try:
        print(f"\n{nombre}")
        ventas_cat = df.groupby("Categoría del Producto")["Precio"].sum()
        for categoria, monto in ventas_cat.items():
            print(f"   {categoria}: ${monto:,.2f}")
    except Exception as e:
        print(f"Hubo un error en {nombre}: {e}")
```

# 3. Calificación promedio de la tienda


```python
print("\n3. Calificación Promedio de cada Tienda")

for nombre, df in [("Tienda 1", tienda), ("Tienda 2", tienda2), ("Tienda 3", tienda3), ("Tienda 4", tienda4)]:
    try:
        promedio = df["Calificación"].mean()
        print(f"{nombre}: {promedio:.2f} ⭐")
    except Exception as e:
        print(f"Hubo un error en {nombre}: {e}")
```

# 4. Productos más y menos vendidos

```python
print("\n4. Productos Más y Menos Vendidos por Tienda")

for nombre, df in [("Tienda 1", tienda), ("Tienda 2", tienda2), ("Tienda 3", tienda3), ("Tienda 4", tienda4)]:
    try:
        print(f"\n{nombre}")
        conteo = df["Producto"].value_counts()
        print("3 Más vendidos:")
        for producto, cantidad in conteo.head(3).items():
            print(f"      {producto}: {cantidad} ventas")
        print("3 Menos vendidos:")
        for producto, cantidad in conteo.tail(3).items():
            print(f"      {producto}: {cantidad} ventas")
    except Exception as e:
        print(f"Hubo un error en {nombre}: {e}")
```

# 5. Envío promedio por tienda

```python
print("\n5. Costo de Envío Promedio por Tienda")

for nombre, df in [("Tienda 1", tienda), ("Tienda 2", tienda2), ("Tienda 3", tienda3), ("Tienda 4", tienda4)]:
    try:
        promedio_envio = df["Costo de envío"].mean()
        print(f"{nombre}: ${promedio_envio:,.2f}")
    except Exception as e:
        print(f"Error en {nombre}: {e}")
```

## 5.1 Tienda 1

```python
import folium
from folium.plugins import MarkerCluster

# Agrupar datos geográficos
geo_tienda1 = tienda.groupby(['lat', 'lon']).agg({
    'Precio': 'sum',
    'Calificación': 'mean'
}).reset_index()

# Crear el mapa centrado en el promedio de coordenadas
m1 = folium.Map(
    location=[geo_tienda1['lat'].mean(), geo_tienda1['lon'].mean()],
    zoom_start=6,
    tiles='CartoDB positron'
)

# Agrupar marcadores
marker_cluster = MarkerCluster().add_to(m1)

# Función para colorear por ventas
def get_color(ventas):
    if ventas > 50000:
        return 'darkred'
    elif ventas > 20000:
        return 'orange'
    else:
        return 'green'

# Añadir marcadores
for i, row in geo_tienda1.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=8,
        color=get_color(row['Precio']),
        fill=True,
        fill_opacity=0.7,
        popup=folium.Popup(f"""
            <b>Ventas:</b> $ {row['Precio']:,.0f}<br>
            <b>Calificación promedio:</b> {row['Calificación']:.2f}
        """, max_width=250)
    ).add_to(marker_cluster)

# Mostrar mapa
m1
```
